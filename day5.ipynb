```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('data.csv')
data.head()
```

```python
X=data['Head Size(cm^3)'].values
Y=data['Brain Weight(grams)'].values
```

```python
mean_x=np.mean(X)
mean_y=np.mean(Y)
```

```python
n=len(X)
```

```python
numer=0
denom=0
for i in range(n):
    numer+=(X[i]-mean_x)*(Y[i]-mean_y)
    denom+=(X[i]-mean_x)**2
b1=numer/denom
b0=mean_y-(b1*mean_x)
print("coefficient")
print(b1,b0)
```

```python
max_x=np.max(X)+100
min_x=np.min(X)-100
```

```python
y=b0+b1*X
plt.plot(X,y,color='#58b970',label='Regression Line')
plt.scatter(X,Y,c='#ef5423',label='Scatter Plot')

```
